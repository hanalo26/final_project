```python
import pandas as pd
```

```python
# 파일 로드
fail_success_data = pd.read_csv('./it_longform_success_data.csv')
classified_data = pd.read_csv('./classified_it_longform_df.csv')

print(f"fail_success_data 행 수: {len(fail_success_data)}")
print(f"classified_data행 수: {len(classified_data)}")
```

```python
# fail_succes_data의 video_id 리스트 추출
fail_success_video_ids = fail_success_data['video_id'].tolist()
print(f"\nfail_succes_data의 video_id 수: {len(fail_success_video_ids)}")

# classified_data에서 fail_succes_data의 video_id와 매칭되는 행 찾기
matched_classified = classified_data[classified_data['video_id'].isin(fail_success_video_ids)]
print(f"매칭된 행 수: {len(matched_classified)}")
```

```python
for col in matched_classified.columns:
    print(col)
```

```python
# 추가할 컬럼에 classification model에서 온 정보라는거 접두사 'cls_'로 표시
matched_classified = matched_classified.rename(
    columns=lambda x: 'cls_' + x if x in ['domain', 'content_type', 'marketing_purpose', 'cta_type', 'is_series', 'is_collaboration', 'reason'] else x
)

for col in matched_classified.columns:
    print(col)
```

```python
# fail_success_data와 matched_classified를 video_id 기준으로 merge
# fail_success_data에서 중복된 video_id가 있을 수 있으므로 확인

print(f"\nfail_success_data에서 중복된 video_id: {fail_success_data['video_id'].duplicated().sum()}")
print(f"classified_data에서 중복된 video_id: {classified_data['video_id'].duplicated().sum()}")

# merge 수행 (left join으로 fail_success_data의 모든 행 유지)
result = fail_success_data.merge(
    matched_classified,
    on='video_id',
    how='left'
)
```

```python
# 결과를 CSV로 저장
output_path = './merged/it_longform_success_data_with_classified.csv'
result.to_csv(output_path, index=False, encoding='utf-8-sig')
print(f"\n 결과 저장: {output_path}")

# 추가 정보
print(f"\n결과 파일 열 수: {len(result.columns)}")
print("\n추가된 분류 관련 열:")
print(result[['video_id', 'title', 'cls_content_type', 'cls_marketing_purpose', 'cls_cta_type', 'cls_is_series', 'cls_is_collaboration']].head(10))
```
