```python
# ========================================
# 라이브러리 로드
# ========================================
from IPython.display import display
from pprint import pprint

import pandas as pd
```

```python
# FnB 도메인에 속하는 기업 영상들 중 숏츠 영상 필터링
df_food = pd.read_csv("../../data/filtered_food_with_redirect.csv", encoding='utf-8')

# domain 컬럼 추가
df_food['domain'] = 'FnB'

print(df_food.shape)
print('-'*30)
pprint(df_food.info())
print('-'*30)
pprint(df_food['verdict'].value_counts())
```

```python
df_food_shorts = df_food[df_food['verdict'] == 'shorts']

print(len(df_food_shorts))
```

```python
# IT 도메인에 속하는 기업 영상들 중 숏츠 영상 필터링
df_it = pd.read_csv("../../data/filtered_it_with_redirect.csv", encoding='utf-8')

# domain 컬럼 추가
df_it['domain'] = 'IT'

print(df_it.shape)
print('-'*30)
pprint(df_it.info())
print('-'*30)
pprint(df_it['verdict'].value_counts())
```

```python
df_it_shorts = df_it[df_it['verdict'] == 'shorts']

print(len(df_it_shorts))
```
