```python
# 데이터 로드
with open("../../outputs/checkpoint_FOOD(배치처리).json", 'r', encoding='utf-8') as f:
    dic_food = json.load(f)
    
pprint(dic_food)
```

```python
df_food = pd.DataFrame(dic_food)

df_food.head(3)
```

```python
# 각 회사마다 수집된 채널의 개수

df_food['channel_count'] = df_food['channels'].apply(len)
```

```python
# 중복된 회사명 제거
df_food_copied = df_food.copy()

# 동일한 회사가 반복적으로 수행된 경우 확인
df_confirm = df_food_copied[df_food_copied.duplicated(subset=['company_name'], keep=False)]
```

```python
# 각 회사별로 channels 결과가 겹치는지 확인
def find_duplicates(main_list):
    # 1단계: 
    flatted = []
    for sub_list in main_list:
        for item in sub_list:
            flatted.append(item['channel_id'])

    # 2단계: 중복 찾기
    duplicates = []
    for item in flatted:
        if flatted.count(item) > 1:
            duplicates.append(item)

    # 3단계: 중복 제거 후 반환
    return list(set(duplicates))
```

```python
df_confirm.groupby('company_name')['channels'].apply(find_duplicates).reset_index()
```

```python
# 회사명이 같으면 동일한 channel_id가 수집된 것을 확인함 -> 회사명이 중복인 경우 첫번째 행만 남기고 삭제

df_food_clean = df_food_copied.drop_duplicates(subset=['company_name'], keep='first')

df_food_clean['company_name'].nunique
```

```python
# channel_count > 0 인 행만 남겨두기
df_food_clean = df_food_clean[df_food_clean['channel_count']>0]

df_food_clean.info()
```

```python
# csv 파일로 저장
df_food_clean.to_csv("../../data/Has_Official_Food-Company-List.csv", encoding='utf-8-sig', index=False)
```
