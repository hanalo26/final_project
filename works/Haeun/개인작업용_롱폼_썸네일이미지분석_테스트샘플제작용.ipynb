### 테스트 데이터 제작

```python
# 테스트 데이터 추출 (성공한 영상)
df_success_fnb_test = df_success_fnb.iloc[501:503]

df_success_fnb_test.head(1)
```

```python
# 테스트 데이터 추출 (실패한 영상)
df_fail_fnb_test = df_fail_fnb.iloc[516:518]

df_fail_fnb_test.head(1)
```

```python
# 테스트 데이터 추출 (성공한 영상)
df_success_it_test = df_success_it.iloc[1500:1502]

df_success_it_test.head(1)
```

```python
# 테스트 데이터 추출 (실패한 영상)
df_fail_it_test = df_fail_it.iloc[12:14]

df_fail_it_test.head(1)
```

```python
df_test = pd.concat([df_success_fnb_test, df_fail_fnb_test, df_success_it_test, df_fail_it_test], ignore_index=True)
```
